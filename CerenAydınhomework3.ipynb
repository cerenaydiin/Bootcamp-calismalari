### ‚ùì Soru: "Age" deƒüi≈ükeninin √ßarpƒ±klƒ±k (skewness) ve basƒ±klƒ±ƒüƒ±nƒ± (kurtosis) hesaplayƒ±n ve yorumlayƒ±n.

### ‚ùì Soru: Eksik veri analizi yaparak eksik verileri heatmap ile g√∂rselle≈ütirin.

### ‚ùì Soru: Matplotlib ile √∂zelle≈ütirilmi≈ü bir bar grafiƒüi olu≈üturun (renk, kalƒ±nlƒ±k, grid, label).

```python
import pandas as pd
import matplotlib.pyplot as plt
fare_means = df.groupby("sex")["fare"].mean()

plt.figure(figsize=(8, 5))
bars = plt.bar(fare_means.index, fare_means.values, color=["hotpink", "salmon"], edgecolor="black", linewidth=2)

for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y + 1, f"{y:.2f}", ha="center")

plt.title("Cinsiyete G√∂re Ortalama Bilet √úcreti")
plt.xlabel("Cinsiyet")
plt.ylabel("Ortalama Fare")
plt.grid(axis="y", linestyle="--", alpha=0.10)
plt.show()

```
